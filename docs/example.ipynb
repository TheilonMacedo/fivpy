## Examples
Calculating a forest inventory using Simple Random Sampling:

```python
import pandas as pd
from fivpy import RandomSample

# Loading the data
data = pd.read_csv('../src/fivpy/data/data_2.csv', sep=';', decimal=',')

inventory_1 = RandomSample(data,
                            unit_area=0.02,
                            sampling_area=11,
                            significance=90,
                            sampling_error=10)

inventory_1.srs_inventory()
```
